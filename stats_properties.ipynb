```python
response = (
    supabase
    .table("properties")
    .select("id, title, price, city")
    .eq("is_published", True)
    .execute()
)

properties = response.data

print(f"{len(properties)} biens publiés récupérés")
properties[:3]  # Aperçu
```

```python
city_counts = defaultdict(int)
city_price_totals = defaultdict(float)

for prop in properties:
    city = prop["city"]
    price = prop["price"] or 0

    city_counts[city] += 1
    city_price_totals[city] += price

print("Calcul terminé")
```

```python
print("Statistiques par ville\n")

for city in city_counts:
    count = city_counts[city]
    avg_price = city_price_totals[city] / count if count > 0 else 0

    print(f"Ville : {city}")
    print(f"Nombre de biens : {count}")
    print(f"Prix moyen : {avg_price:,.2f}")
    print("-" * 40)
```
